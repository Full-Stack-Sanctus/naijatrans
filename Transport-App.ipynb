2. Installing Java

```python
import subprocess
import shutil

def check_java_installed():
    # Check if Java is installed
    java_installed = shutil.which("java")
    return java_installed

def get_installed_java_version():
    try:
        # Get the installed Java version
        result = subprocess.run(["java", "-version"], capture_output=True, text=True, check=True)
        version_line = result.stderr.splitlines()[0]
        # Extract the Java version from the output (e.g., 'java version "17.0.1"')
        version = version_line.split('"')[1]
        return version
    except subprocess.CalledProcessError:
        return None

def install_java(version="21"):
    print(f"Installing OpenJDK {version}...")
    try:
        # Update package lists
        subprocess.run(["sudo", "apt-get", "update"], check=True)

        # Install the specified version of OpenJDK
        subprocess.run(["sudo", "apt-get", "install", "-y", f"openjdk-{version}-jdk"], check=True)

        # Set the new Java version as default
        subprocess.run(["sudo", "update-alternatives", "--set", "java", f"/usr/lib/jvm/java-{version}-openjdk-amd64/bin/java"], check=True)

        # Verify the installation
        subprocess.run(["java", "-version"], check=True)

        print(f"OpenJDK {version} installation completed.")
    except subprocess.CalledProcessError as e:
        print(f"An error occurred during the installation process: {e}")

if __name__ == "__main__":
    desired_java_version = "21"  # Change this to the version you need

    java_installed = check_java_installed()
    if java_installed:
        print("Java is already installed.")
        installed_version = get_installed_java_version()
        print(f"Installed Java version: {installed_version}")

        if not installed_version.startswith(desired_java_version):
            print(f"Installing Java {desired_java_version} because the installed version is {installed_version}.")
            install_java(version=desired_java_version)
        else:
            print(f"Java {desired_java_version} is already installed.")
    else:
        print("Java is not installed.")
        install_java(version=desired_java_version)


    # Check Flutter again
    !flutter doctor
    !java --version
```

```python
import os

# Path to your build.gradle file (replace this with the correct path)
gradle_file_path = '/root/.pub-cache/hosted/pub.dev/babstrap_settings_screen-0.1.7/build.gradle'

def modify_gradle_file():
    # Check if the file exists
    if not os.path.exists(gradle_file_path):
        print(f"Error: {gradle_file_path} does not exist.")
        return

    # Read the file contents
    with open(gradle_file_path, 'r') as file:
        lines = file.readlines()

    # Create a new list to store modified lines
    modified_lines = []
    kotlin_block_found = False
    compile_options_found = False

    # Modify the lines related to jvmTarget or compileOptions
    for line in lines:
        # Find and modify kotlinOptions block
        if "kotlinOptions" in line:
            kotlin_block_found = True

        if kotlin_block_found and "jvmTarget" in line:
            modified_lines.append('        jvmTarget = "21"\n')  # Modify jvmTarget
            kotlin_block_found = False
        else:
            modified_lines.append(line)

        # Find and modify compileOptions block
        if "compileOptions" in line:
            compile_options_found = True

        if compile_options_found and "targetCompatibility" in line:
            modified_lines.append('        targetCompatibility JavaVersion.VERSION_21\n')
            compile_options_found = False
        elif compile_options_found and "sourceCompatibility" in line:
            modified_lines.append('        sourceCompatibility JavaVersion.VERSION_21\n')
        else:
            modified_lines.append(line)

    # Write the modified content back to the build.gradle file
    with open(gradle_file_path, 'w') as file:
        file.writelines(modified_lines)

    print(f"Modified {gradle_file_path} successfully.")

# Call the function
modify_gradle_file()
```

6. Load the keystore Object on build file

```python
# Path to the build.gradle file
build_gradle_path = "/content/my_drive/MyDrive/NaijaTrans/android/app/build.gradle"

# Snippet to load the keystore properties in the build.gradle file
keystore_loading_snippet = """
// Load the keystore properties from the key.properties file
def keystorePropertiesFile = file("/content/my_drive/MyDrive/NaijaTrans/android/config/keystore.properties")
def keystoreProperties = new Properties()

if (keystorePropertiesFile.exists()) {
    keystorePropertiesFile.withReader { reader ->
        keystoreProperties.load(reader)
    }
} else {
    throw new GradleException("key.properties file not found. Ensure the file is located in the project's root directory.")
}
"""

def modify_build_gradle():
    try:
        # Read the content of build.gradle
        with open(build_gradle_path, 'r') as file:
            content = file.readlines()

        # Check if the keystore loading snippet is already present
        if "def keystorePropertiesFile" in ''.join(content):
            print("Keystore properties loading is already configured in build.gradle.")
            return

        # Find the position of the android block
        android_block_index = None
        for i, line in enumerate(content):
            if line.strip().startswith("android {"):
                android_block_index = i
                break

        # If android block is found, insert the keystore loading snippet right before it
        if android_block_index is not None:
            content.insert(android_block_index, keystore_loading_snippet + "\n")
        else:
            print("Error: 'android {}' block not found in the build.gradle file.")
            return

        # Write the modified content back to build.gradle
        with open(build_gradle_path, 'w') as file:
            file.writelines(content)

        print(f"Successfully updated {build_gradle_path} with keystore loading configuration before the android block.")

    except Exception as e:
        print(f"Error modifying build.gradle: {e}")

# Call the function to modify build.gradle
modify_build_gradle()
```
